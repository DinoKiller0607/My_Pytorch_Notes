## PyTorch Datatypes
**Note:** Tensor datatypes are one of the three big errors you might run into with PyTorch and deep learning:
1. Tensors are not the right data type
2. Tensors are not the right shape
3. Tensors are not on the right device

```python
# Float32 tensor
float_32_tensor = torch.tensor([3.0, 6.0, 7.0], # Even if datatype is set to None, datatype will still be float32
                               dtype=None,      # data type of the tensor
                               device=None,     # what device the tensor is on
                               requires_grad=False) # whether to track gradients
print(f"float_32_tensor = {float_32_tensor}"
      f"\nfloat_32_tensor.dtype = {float_32_tensor.dtype}")
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
print(f"float_16_tensor = {float_16_tensor}")
```

```python
unknown_tensor = float_32_tensor * float_16_tensor      # pytorch will automatically convert the datatype to the one with more information
print(f"unknown_tensor = {unknown_tensor}\nunknown_tensor.dtype = {unknown_tensor.dtype}")
```

```python
int_32_tensor = torch.tensor([3, 6, 9], dtype=torch.int32)
print(f"int_32_tensor = {int_32_tensor}\nint_32_tensor.dtype = {int_32_tensor.dtype}")
```

## Matrix Multiplication
Two main ways of performing multiplication in neural networks and deep learning:
1. Element-wise Multiplication (Scalar Multiplication)
2. Matrix Multiplication

Check out the http://matrixmultiplication.xyz/ page for intuition on how **matrix multiplication** works.

There are two main rules that performing matrix multiplication needs to satisfy:
1. The **inner dimensions** must match:
    * `(3, 2) @ (2, 3)` won't work
    * `(2, 3) @ (3, 2)` will work
    * `(3, 2) @ (2, 3)` will work
2. The resulting matrix will have the shape of the **outer dimensions**
    * `(3, 2) @ (2, 3)` -> `(3, 3)`
    * `(2, 3) @ (3, 5)` -> `(2, 5)`

```python
# Element-wise multiplication
print(f"{tensor} * {tensor}"
      f"Equals: {tensor * tensor}")
```

```python
# Matrix multiplication
print(f"Matrix Multiplication: {tensor} @ {tensor}"
      f" Equals: {torch.matmul(tensor, tensor)}")    # Can also use tensor @ tensor
```

```python
%%time
value = 0
for i in range(len(tensor)):
    value += tensor[i] * tensor[i]
value
```

```python
%%time
torch.matmul(tensor, tensor)
```

## One of the most common errors in deep learning: shape errors


```python
# Shapes for matrix multiplication
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])
tensor_B = torch.tensor([[7, 8],
                         [9, 10],
                         [9, 12]])
torch.matmul(tensor_A, tensor_B)
```

To fix our tensor shape issues, we can use a **transpose**.
A **transpose** switches the axes or shape of a given tensor.


```python
tensor_B, tensor_B.shape
```

```python
tensor_B.T, tensor_B.T.shape
```

```python
print(f"Original shapes: tensor_A =  {tensor_A.shape}, tensor_B = {tensor_B.shape}")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B = {tensor_B.T.shape}")
print("Result of matrix multiplication:")
output = torch.matmul(tensor_A, tensor_B.T)
print(f"Output = {output}"
      f"\nOutput shape = {output.shape}")
```

## Reshaping, stacking, squeezing, and unsqueezing tensors
* Reshaping — reshapes an input tensor to a defined shape
* View — Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking — combine multiple tensors on top of each other (vstack) or side by side (hstack)
* Squeeze — removes all `1` dimensions from a tensor
* Unsqueeze — add a `1` dimension to a target vector
* Permute — Return a view of the input with dimensions permuted (swapped) in a certain way.

```python
# Let's create a tensor
import torch
x = torch.arange(1., 10.,)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 9)
z, z.shape
```

```python
# Changing z changes x (because a view of a tensor shares the same memory as the original input)
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked
```

```python
y = torch.rand(2, 1, 2, 1, 2)
y, y.shape
```

```python
# torch.squeeze() - removes all single dimensions from a target tensor
print(f"Previous tensor: {y}"
      f"\nPrevious shape: {y.shape}")
```

```python
y_squeezed = torch.squeeze(y)
print(f"New tensor: {y_squeezed}"
      f"\nNew shape: {y_squeezed.shape}")
```

```python
# torch.unsqueeze() – adds a single dimension to a target tensor at a specific diam (dimension)
print(f"Previous target: {y_squeezed}"
      f"\nPrevious shape: {y_squeezed.shape}")
# Add an extra dimension with unsqueeze
y_unsqueezed = y_squeezed.unsqueeze(dim=1)
print(f"New tensor: {y_unsqueezed}"
      f"\nNew Shape: {y_unsqueezed.shape}")
```

```python
# torch.permute() — rearranges the dimensions of a target tensor in a specified order
x_original = torch.rand(size=(224, 224, 3)) # [height, width, colour_channels]

# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) #shifts axis 0->1, 1->2, 2->0
print(f"Previous shape: {x_original.shape}"
      f"\nNew shape: {x_permuted.shape}")
```

```python
x_original[0, 0, 0] = 786
x_original[0, 0, 0], x_permuted[0, 0, 0]
```

## PyTorch and NumPy
NumPy is a popular scientific Python numerical library.
And because of this, PyTorch has functionality to react with it.
* Data in NumPy, want in PyTorch tensor → `torch.from_numpy(ndarray)`
* PyTorch tensor → NumPy → `torch.Tensor.numpy()`
### NumPy Array to Tensor

```python
import torch
import numpy as np
array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

```python
# Default data type for numpy float array
array.dtype
```

```python
# Default data type for a float tensor in pytorch
torch.arange(1.0, 8.0).dtype
```

**WARNING:** When converting from NumPy → PyTorch, PyTorch reflects NumPy's default datatype of float64 unless specified otherwise: like,
`tensor = torch.from_numpy(array).type(tortch.float32)`.
Now if we change the value of `array`, what will this do to `tensor`?

```python
array = array + 1
array, tensor
```

As seen above, the pytorch tensor is created in memory when using from_numpy(), so its value does not change
### Tensor to NumPy Array

## Reproducibility (Trying to take random out of randomness)
In short, how a neural network learns:
`start with random numbers → tensor operations → update random numbers to try and make them better representations of data
→ again, again and, again...`

To reduce the randomness in neural networks and pytorch, comes the concept of **random seeds.**
Essentially, what the random seed does is that it adds "flavour" to the randomness.

```python
# Create two random tensors
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)
print(f"random_tensor_A : {random_tensor_A}")
print(f"random_tensor_B : {random_tensor_B}")
print(random_tensor_A == random_tensor_B)
```

```python
# Let's make some random but reproducible tensors
import torch
# Set the random seed
RANDOM_STATE = 42
torch.manual_seed(RANDOM_STATE)
random_tensor_C = torch.rand(3, 4)
torch.manual_seed(RANDOM_STATE)
random_tensor_D = torch.rand(3, 4)
print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```

In pytorch, you have to set the random seed everytime you call the `rand` function.

Since PyTorch can run on both a GPU or a CPU, curate/customize your code to be device-agnostic, meaning:

```python
# Set up device-agnostic code
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
# Check the number of GPUs
torch.cuda.device_count()
```

### Putting tensors (and models on the GPU)
The reason we want to put our tensors on a GPU is that using a GPU results in faster computation.

```python
# Create a tensor (default is on the CPU)
tensor = torch.tensor([1, 2, 3])
print(tensor, tensor.device)    # tensor not on gpu
# Move the tensor to the GPU (if available)
tensor_on_gpu = tensor.to(device)
print(tensor_on_gpu, tensor_on_gpu.device)
```

### Putting tensors back to the CPU

```python
# If tensor is on GPU, can't transform it to numpy
try:
    tensor_on_gpu.numpy()
except Exception as e:
    print(f"Error Type: {type(e).__name__} | Message: {e}")
```

```python
# To fix the GPU tensor with the NumPy issue, we can first set it to the CPU
tensor_back_on_cpu = tensor_on_gpu.cpu().numpy()
tensor_back_on_cpu, tensor_back_on_cpu.device
```
